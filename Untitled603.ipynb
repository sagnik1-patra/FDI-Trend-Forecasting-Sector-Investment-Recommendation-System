```python
import os
import pandas as pd
import json
import yaml

# ================================
# 📂 File Paths
# ================================
input_path = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System\FDI-2000-01_to_2015-16_1.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System"

# ================================
# 📥 Load Dataset
# ================================
df = pd.read_csv(input_path)

# ================================
# 🧹 Basic Cleaning
# ================================
# Remove unnamed columns if present
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Fill missing values
df.fillna(0, inplace=True)

# Standardize column names
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ================================
# 🔄 Transform (Optional but useful)
# Convert wide format → long format
# ================================
if "Sector" in df.columns:
    df_long = df.melt(id_vars=["Sector"], var_name="Year", value_name="FDI")
else:
    # fallback if column name is different
    df_long = df.melt(var_name="Year", value_name="FDI")

# Convert Year column properly
df_long["Year"] = df_long["Year"].astype(str)

# ================================
# 💾 Save Files
# ================================

# 1. HDF5 (.h5)
h5_path = os.path.join(output_dir, "fdi_data.h5")
df_long.to_hdf(h5_path, key='df', mode='w')

# 2. Pickle (.pkl)
pkl_path = os.path.join(output_dir, "fdi_data.pkl")
df_long.to_pickle(pkl_path)

# 3. JSON (.json)
json_path = os.path.join(output_dir, "fdi_data.json")
df_long.to_json(json_path, orient="records", indent=4)

# 4. YAML (.yaml)
yaml_path = os.path.join(output_dir, "fdi_data.yaml")
with open(yaml_path, "w") as file:
    yaml.dump(df_long.to_dict(orient="records"), file, default_flow_style=False)

# ================================
# ✅ Done
# ================================
print("Files successfully saved:")
print(f"HDF5: {h5_path}")
print(f"Pickle: {pkl_path}")
print(f"JSON: {json_path}")
print(f"YAML: {yaml_path}")
```
