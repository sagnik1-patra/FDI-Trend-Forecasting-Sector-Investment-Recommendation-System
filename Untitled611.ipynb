```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# ================================
# 📂 Paths
# ================================
input_path = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System\FDI-2000-01_to_2015-16_1.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System"
os.makedirs(output_dir, exist_ok=True)

# ================================
# 📥 Load & Clean
# ================================
df = pd.read_csv(input_path)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ================================
# 🔄 Transform
# ================================
df_long = df.melt(id_vars=["Sector"], var_name="Year", value_name="FDI")

# Safe Year Fix
df_long["Year"] = df_long["Year"].astype(str).str.extract(r'(\d{4})')
df_long = df_long.dropna(subset=["Year"])
df_long["Year"] = df_long["Year"].astype(int)

# Encode Sector
df_long["Sector_Code"] = df_long["Sector"].astype('category').cat.codes

# ================================
# 🎯 Features
# ================================
X = df_long[["Year", "Sector_Code"]].values
y = df_long["FDI"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# ================================
# 🐋 WOA REGRESSOR
# ================================
class WOARegressor:
    def __init__(self, n_whales=30, iterations=60):
        self.n_whales = n_whales
        self.iterations = iterations

    def fitness(self, w, X, y):
        return mean_squared_error(y, X.dot(w))

    def fit(self, X, y):
        dim = X.shape[1]

        whales = np.random.randn(self.n_whales, dim)
        fitness = np.array([self.fitness(w, X, y) for w in whales])

        best_idx = np.argmin(fitness)
        best = whales[best_idx]

        for t in range(self.iterations):
            a = 2 - t * (2 / self.iterations)

            for i in range(self.n_whales):
                r1, r2 = np.random.rand(), np.random.rand()
                A = 2 * a * r1 - a
                C = 2 * r2
                p = np.random.rand()

                if p < 0.5:
                    if abs(A) < 1:
                        D = abs(C * best - whales[i])
                        whales[i] = best - A * D
                    else:
                        rand_whale = whales[np.random.randint(self.n_whales)]
                        D = abs(C * rand_whale - whales[i])
                        whales[i] = rand_whale - A * D
                else:
                    D = abs(best - whales[i])
                    l = np.random.uniform(-1, 1)
                    whales[i] = D * np.exp(1 * l) * np.cos(2 * np.pi * l) + best

                fitness[i] = self.fitness(whales[i], X, y)

            best_idx = np.argmin(fitness)
            best = whales[best_idx]

        self.best_w = best

    def predict(self, X):
        return X.dot(self.best_w)

# ================================
# 🤖 Train Model
# ================================
model = WOARegressor()
model.fit(X_train, y_train)

preds = model.predict(X_test)

# ================================
# 📊 Metrics
# ================================
r2 = r2_score(y_test, preds)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

print("\n===== WOA MODEL RESULTS =====")
print(f"R2 Score : {r2:.4f}")
print(f"MAE      : {mae:.4f}")
print(f"RMSE     : {rmse:.4f}")

# ================================
# 💾 Save Results CSV
# ================================
results_df = pd.DataFrame({
    "Model": ["WOA"],
    "R2": [r2],
    "MAE": [mae],
    "RMSE": [rmse]
})
results_df.to_csv(os.path.join(output_dir, "woa_model_results.csv"), index=False)

# ================================
# 📊 Accuracy Graph
# ================================
plt.figure()
plt.bar(["WOA"], [r2])
plt.title("WOA Accuracy (R2)")
plt.savefig(os.path.join(output_dir, "woa_accuracy_graph.png"))
plt.show()

# ================================
# 📊 Comparison Graph
# ================================
results_df.set_index("Model").plot(kind="bar")
plt.title("WOA Model Comparison")
plt.savefig(os.path.join(output_dir, "woa_comparison_graph.png"))
plt.show()

# ================================
# 🔥 Heatmap
# ================================
plt.figure()
sns.heatmap(results_df.set_index("Model"), annot=True, cmap="coolwarm")
plt.title("WOA Heatmap")
plt.savefig(os.path.join(output_dir, "woa_heatmap.png"))
plt.show()

# ================================
# 📁 Prediction CSV
# ================================
pred_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": preds
})
pred_df.to_csv(os.path.join(output_dir, "woa_predictions.csv"), index=False)

# ================================
# 📉 Prediction Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="WOA Predicted")
plt.legend()
plt.title("WOA Prediction Graph")
plt.savefig(os.path.join(output_dir, "woa_prediction_graph.png"))
plt.show()

# ================================
# 📈 Result Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("WOA Result Graph")
plt.savefig(os.path.join(output_dir, "woa_result_graph.png"))
plt.show()

# ================================
# 🔮 Future Prediction
# ================================
future_years = np.arange(2017, 2025)
future_data = []

for sector in df_long["Sector_Code"].unique():
    for year in future_years:
        future_data.append([year, sector])

future_df = pd.DataFrame(future_data, columns=["Year", "Sector_Code"])
future_preds = model.predict(future_df.values)

future_df["Predicted_FDI"] = future_preds
future_df.to_csv(os.path.join(output_dir, "woa_future_predictions.csv"), index=False)

# ================================
# 📉 Future Graph
# ================================
plt.figure()
plt.plot(future_preds)
plt.title("WOA Future Prediction Graph")
plt.savefig(os.path.join(output_dir, "woa_future_graph.png"))
plt.show()

# ================================
# ✅ DONE
# ================================
print("\n✅ WOA pipeline completed successfully!")
print("📁 All files saved with 'woa_' prefix")
```
