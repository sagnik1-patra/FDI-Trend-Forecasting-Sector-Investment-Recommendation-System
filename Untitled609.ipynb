```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# ================================
# 📂 Paths
# ================================
input_path = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System\FDI-2000-01_to_2015-16_1.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System"
os.makedirs(output_dir, exist_ok=True)

# ================================
# 📥 Load & Clean
# ================================
df = pd.read_csv(input_path)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ================================
# 🔄 Transform
# ================================
df_long = df.melt(id_vars=["Sector"], var_name="Year", value_name="FDI")

df_long["Year"] = df_long["Year"].astype(str).str.extract(r'(\d{4})')
df_long = df_long.dropna(subset=["Year"])
df_long["Year"] = df_long["Year"].astype(int)

df_long["Sector_Code"] = df_long["Sector"].astype('category').cat.codes

# ================================
# 🎯 Features
# ================================
X = df_long[["Year", "Sector_Code"]].values
y = df_long["FDI"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# ================================
# 🎼 HARMONY SEARCH REGRESSOR
# ================================
class HSARegressor:
    def __init__(self, HMS=30, HMCR=0.9, PAR=0.3, bw=0.01, iterations=60):
        self.HMS = HMS        # Harmony Memory Size
        self.HMCR = HMCR      # Harmony Memory Consideration Rate
        self.PAR = PAR        # Pitch Adjustment Rate
        self.bw = bw          # Bandwidth
        self.iterations = iterations

    def fit(self, X, y):
        dim = X.shape[1]

        # Initialize Harmony Memory
        HM = np.random.randn(self.HMS, dim)
        fitness = np.array([mean_squared_error(y, X.dot(h)) for h in HM])

        for _ in range(self.iterations):
            new_harmony = np.zeros(dim)

            for d in range(dim):
                if np.random.rand() < self.HMCR:
                    idx = np.random.randint(0, self.HMS)
                    new_harmony[d] = HM[idx, d]

                    if np.random.rand() < self.PAR:
                        new_harmony[d] += self.bw * np.random.randn()
                else:
                    new_harmony[d] = np.random.randn()

            new_fitness = mean_squared_error(y, X.dot(new_harmony))

            worst_idx = np.argmax(fitness)
            if new_fitness < fitness[worst_idx]:
                HM[worst_idx] = new_harmony
                fitness[worst_idx] = new_fitness

        self.best_w = HM[np.argmin(fitness)]

    def predict(self, X):
        return X.dot(self.best_w)

# ================================
# 🤖 Train Model
# ================================
model = HSARegressor()
model.fit(X_train, y_train)

preds = model.predict(X_test)

# ================================
# 📊 Metrics
# ================================
r2 = r2_score(y_test, preds)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

print("\n===== HSA MODEL RESULTS =====")
print(f"R2 Score : {r2:.4f}")
print(f"MAE      : {mae:.4f}")
print(f"RMSE     : {rmse:.4f}")

# ================================
# 💾 Save Results CSV
# ================================
results_df = pd.DataFrame({
    "Model": ["HSA"],
    "R2": [r2],
    "MAE": [mae],
    "RMSE": [rmse]
})
results_df.to_csv(os.path.join(output_dir, "hsa_model_results.csv"), index=False)

# ================================
# 📊 Accuracy Graph
# ================================
plt.figure()
plt.bar(["HSA"], [r2])
plt.title("HSA Accuracy (R2)")
plt.savefig(os.path.join(output_dir, "hsa_accuracy_graph.png"))
plt.show()

# ================================
# 📊 Comparison Graph
# ================================
results_df.set_index("Model").plot(kind="bar")
plt.title("HSA Model Comparison")
plt.savefig(os.path.join(output_dir, "hsa_comparison_graph.png"))
plt.show()

# ================================
# 🔥 Heatmap
# ================================
plt.figure()
sns.heatmap(results_df.set_index("Model"), annot=True, cmap="coolwarm")
plt.title("HSA Heatmap")
plt.savefig(os.path.join(output_dir, "hsa_heatmap.png"))
plt.show()

# ================================
# 📁 Prediction CSV
# ================================
pred_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": preds
})
pred_df.to_csv(os.path.join(output_dir, "hsa_predictions.csv"), index=False)

# ================================
# 📉 Prediction Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="HSA Predicted")
plt.legend()
plt.title("HSA Prediction Graph")
plt.savefig(os.path.join(output_dir, "hsa_prediction_graph.png"))
plt.show()

# ================================
# 📈 Result Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("HSA Result Graph")
plt.savefig(os.path.join(output_dir, "hsa_result_graph.png"))
plt.show()

# ================================
# 🔮 Future Prediction
# ================================
future_years = np.arange(2017, 2025)
future_data = []

for sector in df_long["Sector_Code"].unique():
    for year in future_years:
        future_data.append([year, sector])

future_df = pd.DataFrame(future_data, columns=["Year", "Sector_Code"])
future_preds = model.predict(future_df.values)

future_df["Predicted_FDI"] = future_preds
future_df.to_csv(os.path.join(output_dir, "hsa_future_predictions.csv"), index=False)

# ================================
# 📉 Future Graph
# ================================
plt.figure()
plt.plot(future_preds)
plt.title("HSA Future Prediction Graph")
plt.savefig(os.path.join(output_dir, "hsa_future_graph.png"))
plt.show()

# ================================
# ✅ DONE
# ================================
print("\n✅ HSA pipeline completed successfully!")
print("📁 All files saved with 'hsa_' prefix")
```
