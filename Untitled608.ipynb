```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# ================================
# 📂 Paths
# ================================
input_path = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System\FDI-2000-01_to_2015-16_1.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System"
os.makedirs(output_dir, exist_ok=True)

# ================================
# 📥 Load & Clean
# ================================
df = pd.read_csv(input_path)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ================================
# 🔄 Transform
# ================================
df_long = df.melt(id_vars=["Sector"], var_name="Year", value_name="FDI")

df_long["Year"] = df_long["Year"].astype(str).str.extract(r'(\d{4})')
df_long = df_long.dropna(subset=["Year"])
df_long["Year"] = df_long["Year"].astype(int)

df_long["Sector_Code"] = df_long["Sector"].astype('category').cat.codes

# ================================
# 🎯 Features
# ================================
X = df_long[["Year", "Sector_Code"]].values
y = df_long["FDI"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# ================================
# 🧠 BAT ALGORITHM REGRESSOR
# ================================
class BatRegressor:
    def __init__(self, n_bats=25, iterations=60, fmin=0, fmax=2, alpha=0.9, gamma=0.9):
        self.n_bats = n_bats
        self.iterations = iterations
        self.fmin = fmin
        self.fmax = fmax
        self.alpha = alpha
        self.gamma = gamma

    def fit(self, X, y):
        dim = X.shape[1]

        # Initialize
        bats = np.random.randn(self.n_bats, dim)
        velocities = np.zeros((self.n_bats, dim))
        freq = np.zeros(self.n_bats)
        loudness = np.ones(self.n_bats)
        pulse_rate = np.random.rand(self.n_bats)

        fitness = np.array([mean_squared_error(y, X.dot(b)) for b in bats])
        best = bats[np.argmin(fitness)]

        for t in range(self.iterations):
            for i in range(self.n_bats):
                freq[i] = self.fmin + (self.fmax - self.fmin) * np.random.rand()
                velocities[i] += (bats[i] - best) * freq[i]
                bats[i] += velocities[i]

                # Local search
                if np.random.rand() > pulse_rate[i]:
                    bats[i] = best + 0.001 * np.random.randn(dim)

                new_fitness = mean_squared_error(y, X.dot(bats[i]))

                # Accept solution
                if (new_fitness < fitness[i]) and (np.random.rand() < loudness[i]):
                    fitness[i] = new_fitness
                    loudness[i] *= self.alpha
                    pulse_rate[i] = pulse_rate[i] * (1 - np.exp(-self.gamma * t))

            best = bats[np.argmin(fitness)]

        self.best_w = best

    def predict(self, X):
        return X.dot(self.best_w)

# ================================
# 🤖 Train Model
# ================================
model = BatRegressor()
model.fit(X_train, y_train)

preds = model.predict(X_test)

# ================================
# 📊 Metrics
# ================================
r2 = r2_score(y_test, preds)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

print("\n===== BAT MODEL RESULTS =====")
print(f"R2 Score : {r2:.4f}")
print(f"MAE      : {mae:.4f}")
print(f"RMSE     : {rmse:.4f}")

# ================================
# 💾 Save Results CSV
# ================================
results_df = pd.DataFrame({
    "Model": ["BA"],
    "R2": [r2],
    "MAE": [mae],
    "RMSE": [rmse]
})
results_df.to_csv(os.path.join(output_dir, "ba_model_results.csv"), index=False)

# ================================
# 📊 Accuracy Graph
# ================================
plt.figure()
plt.bar(["BA"], [r2])
plt.title("BA Accuracy (R2)")
plt.savefig(os.path.join(output_dir, "ba_accuracy_graph.png"))
plt.show()

# ================================
# 📊 Comparison Graph
# ================================
results_df.set_index("Model").plot(kind="bar")
plt.title("BA Model Comparison")
plt.savefig(os.path.join(output_dir, "ba_comparison_graph.png"))
plt.show()

# ================================
# 🔥 Heatmap
# ================================
plt.figure()
sns.heatmap(results_df.set_index("Model"), annot=True, cmap="coolwarm")
plt.title("BA Heatmap")
plt.savefig(os.path.join(output_dir, "ba_heatmap.png"))
plt.show()

# ================================
# 📁 Prediction CSV
# ================================
pred_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": preds
})
pred_df.to_csv(os.path.join(output_dir, "ba_predictions.csv"), index=False)

# ================================
# 📉 Prediction Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="BA Predicted")
plt.legend()
plt.title("BA Prediction Graph")
plt.savefig(os.path.join(output_dir, "ba_prediction_graph.png"))
plt.show()

# ================================
# 📈 Result Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("BA Result Graph")
plt.savefig(os.path.join(output_dir, "ba_result_graph.png"))
plt.show()

# ================================
# 🔮 Future Prediction
# ================================
future_years = np.arange(2017, 2025)
future_data = []

for sector in df_long["Sector_Code"].unique():
    for year in future_years:
        future_data.append([year, sector])

future_df = pd.DataFrame(future_data, columns=["Year", "Sector_Code"])
future_preds = model.predict(future_df.values)

future_df["Predicted_FDI"] = future_preds
future_df.to_csv(os.path.join(output_dir, "ba_future_predictions.csv"), index=False)

# ================================
# 📉 Future Graph
# ================================
plt.figure()
plt.plot(future_preds)
plt.title("BA Future Prediction Graph")
plt.savefig(os.path.join(output_dir, "ba_future_graph.png"))
plt.show()

# ================================
# ✅ DONE
# ================================
print("\n✅ BAT Algorithm pipeline completed!")
print("📁 All outputs saved with 'ba_' prefix")
```
