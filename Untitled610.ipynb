```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# ================================
# 📂 Paths
# ================================
input_path = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System\FDI-2000-01_to_2015-16_1.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System"
os.makedirs(output_dir, exist_ok=True)

# ================================
# 📥 Load & Clean
# ================================
df = pd.read_csv(input_path)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ================================
# 🔄 Transform
# ================================
df_long = df.melt(id_vars=["Sector"], var_name="Year", value_name="FDI")

df_long["Year"] = df_long["Year"].astype(str).str.extract(r'(\d{4})')
df_long = df_long.dropna(subset=["Year"])
df_long["Year"] = df_long["Year"].astype(int)

df_long["Sector_Code"] = df_long["Sector"].astype('category').cat.codes

# ================================
# 🎯 Features
# ================================
X = df_long[["Year", "Sector_Code"]].values
y = df_long["FDI"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# ================================
# 🐜 ANT LION OPTIMIZER
# ================================
class ALOARegressor:
    def __init__(self, n_agents=30, iterations=60):
        self.n_agents = n_agents
        self.iterations = iterations

    def fitness(self, w, X, y):
        return mean_squared_error(y, X.dot(w))

    def random_walk(self, dim):
        return np.cumsum(2 * (np.random.rand(dim) > 0.5) - 1)

    def fit(self, X, y):
        dim = X.shape[1]

        # Initialize ants and antlions
        ants = np.random.randn(self.n_agents, dim)
        antlions = np.random.randn(self.n_agents, dim)

        fitness_antlions = np.array([self.fitness(a, X, y) for a in antlions])
        elite = antlions[np.argmin(fitness_antlions)]

        for t in range(self.iterations):
            new_ants = []

            for i in range(self.n_agents):
                # Select antlion via roulette wheel
                probs = 1 / (fitness_antlions + 1e-10)
                probs /= probs.sum()
                idx = np.random.choice(range(self.n_agents), p=probs)

                walk = self.random_walk(dim)
                new_ant = (antlions[idx] + elite) / 2 + 0.01 * walk

                new_ants.append(new_ant)

            ants = np.array(new_ants)

            # Update antlions
            for i in range(self.n_agents):
                if self.fitness(ants[i], X, y) < fitness_antlions[i]:
                    antlions[i] = ants[i]
                    fitness_antlions[i] = self.fitness(ants[i], X, y)

            elite = antlions[np.argmin(fitness_antlions)]

        self.best_w = elite

    def predict(self, X):
        return X.dot(self.best_w)

# ================================
# 🤖 Train Model
# ================================
model = ALOARegressor()
model.fit(X_train, y_train)

preds = model.predict(X_test)

# ================================
# 📊 Metrics
# ================================
r2 = r2_score(y_test, preds)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

print("\n===== ALOA MODEL RESULTS =====")
print(f"R2 Score : {r2:.4f}")
print(f"MAE      : {mae:.4f}")
print(f"RMSE     : {rmse:.4f}")

# ================================
# 💾 Save Results CSV
# ================================
results_df = pd.DataFrame({
    "Model": ["ALOA"],
    "R2": [r2],
    "MAE": [mae],
    "RMSE": [rmse]
})
results_df.to_csv(os.path.join(output_dir, "aloa_model_results.csv"), index=False)

# ================================
# 📊 Accuracy Graph
# ================================
plt.figure()
plt.bar(["ALOA"], [r2])
plt.title("ALOA Accuracy (R2)")
plt.savefig(os.path.join(output_dir, "aloa_accuracy_graph.png"))
plt.show()

# ================================
# 📊 Comparison Graph
# ================================
results_df.set_index("Model").plot(kind="bar")
plt.title("ALOA Model Comparison")
plt.savefig(os.path.join(output_dir, "aloa_comparison_graph.png"))
plt.show()

# ================================
# 🔥 Heatmap
# ================================
plt.figure()
sns.heatmap(results_df.set_index("Model"), annot=True, cmap="coolwarm")
plt.title("ALOA Heatmap")
plt.savefig(os.path.join(output_dir, "aloa_heatmap.png"))
plt.show()

# ================================
# 📁 Prediction CSV
# ================================
pred_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": preds
})
pred_df.to_csv(os.path.join(output_dir, "aloa_predictions.csv"), index=False)

# ================================
# 📉 Prediction Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="ALOA Predicted")
plt.legend()
plt.title("ALOA Prediction Graph")
plt.savefig(os.path.join(output_dir, "aloa_prediction_graph.png"))
plt.show()

# ================================
# 📈 Result Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("ALOA Result Graph")
plt.savefig(os.path.join(output_dir, "aloa_result_graph.png"))
plt.show()

# ================================
# 🔮 Future Prediction
# ================================
future_years = np.arange(2017, 2025)
future_data = []

for sector in df_long["Sector_Code"].unique():
    for year in future_years:
        future_data.append([year, sector])

future_df = pd.DataFrame(future_data, columns=["Year", "Sector_Code"])
future_preds = model.predict(future_df.values)

future_df["Predicted_FDI"] = future_preds
future_df.to_csv(os.path.join(output_dir, "aloa_future_predictions.csv"), index=False)

# ================================
# 📉 Future Graph
# ================================
plt.figure()
plt.plot(future_preds)
plt.title("ALOA Future Prediction Graph")
plt.savefig(os.path.join(output_dir, "aloa_future_graph.png"))
plt.show()

# ================================
# ✅ DONE
# ================================
print("\n✅ ALOA pipeline completed successfully!")
print("📁 All files saved with 'aloa_' prefix")
```
