```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split

# ================================
# 📂 Paths
# ================================
input_path = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System\FDI-2000-01_to_2015-16_1.csv"
output_dir = r"C:\Users\NXTWAVE\Downloads\FDI Trend Forecasting & Sector Investment Recommendation System"

os.makedirs(output_dir, exist_ok=True)

# ================================
# 📥 Load & Clean
# ================================
df = pd.read_csv(input_path)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# ================================
# 🔄 Melt
# ================================
df_long = df.melt(id_vars=["Sector"], var_name="Year", value_name="FDI")

# ✅ FIX: Safe year extraction
df_long["Year"] = df_long["Year"].astype(str).str.extract(r'(\d{4})')

# Drop invalid rows
df_long = df_long.dropna(subset=["Year"])

# Convert to int
df_long["Year"] = df_long["Year"].astype(int)

# Encode sector
df_long["Sector_Code"] = df_long["Sector"].astype('category').cat.codes

# ================================
# 🎯 Features
# ================================
X = df_long[["Year", "Sector_Code"]].values
y = df_long["FDI"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# ================================
# 🧠 AIS MODEL (Clonal Selection)
# ================================
class AISRegressor:
    def __init__(self, pop_size=20, generations=50, mutation_rate=0.1):
        self.pop_size = pop_size
        self.generations = generations
        self.mutation_rate = mutation_rate

    def init_population(self, dim):
        return [np.random.randn(dim) for _ in range(self.pop_size)]

    def fitness(self, w, X, y):
        preds = X.dot(w)
        return -mean_squared_error(y, preds)

    def mutate(self, w):
        return w + self.mutation_rate * np.random.randn(len(w))

    def fit(self, X, y):
        dim = X.shape[1]
        population = self.init_population(dim)

        for _ in range(self.generations):
            scores = [(self.fitness(w, X, y), w) for w in population]
            scores.sort(reverse=True)

            best = [w for _, w in scores[:5]]

            clones = []
            for w in best:
                for _ in range(5):
                    clones.append(self.mutate(w))

            population = best + clones

        self.best_w = scores[0][1]

    def predict(self, X):
        return X.dot(self.best_w)

# ================================
# 🤖 Train AIS
# ================================
model = AISRegressor()
model.fit(X_train, y_train)

preds = model.predict(X_test)

# ================================
# 📊 Metrics
# ================================
r2 = r2_score(y_test, preds)
mae = mean_absolute_error(y_test, preds)
rmse = np.sqrt(mean_squared_error(y_test, preds))

results_df = pd.DataFrame({
    "Model": ["AIS"],
    "R2": [r2],
    "MAE": [mae],
    "RMSE": [rmse]
})

# Save results CSV
results_df.to_csv(os.path.join(output_dir, "ais_model_results.csv"), index=False)

# ================================
# 📈 Accuracy Graph
# ================================
plt.figure()
plt.bar(["AIS"], [r2])
plt.title("AIS Accuracy (R2)")
plt.savefig(os.path.join(output_dir, "ais_accuracy_graph.png"))
plt.close()

# ================================
# 📊 Comparison Graph
# ================================
results_df.set_index("Model").plot(kind="bar")
plt.title("AIS Model Metrics")
plt.savefig(os.path.join(output_dir, "ais_comparison_graph.png"))
plt.close()

# ================================
# 🔥 Heatmap
# ================================
plt.figure()
sns.heatmap(results_df.set_index("Model"), annot=True, cmap="coolwarm")
plt.title("AIS Heatmap")
plt.savefig(os.path.join(output_dir, "ais_heatmap.png"))
plt.close()

# ================================
# 📁 Prediction CSV
# ================================
pred_df = pd.DataFrame({
    "Actual": y_test,
    "Predicted": preds
})
pred_df.to_csv(os.path.join(output_dir, "ais_predictions.csv"), index=False)

# ================================
# 📉 Prediction Graph
# ================================
plt.figure()
plt.plot(y_test, label="Actual")
plt.plot(preds, label="AIS Predicted")
plt.legend()
plt.title("AIS Prediction Graph")
plt.savefig(os.path.join(output_dir, "ais_prediction_graph.png"))
plt.close()

# ================================
# 🔮 Future Prediction
# ================================
future_years = np.arange(2017, 2025)
future_data = []

for sector in df_long["Sector_Code"].unique():
    for year in future_years:
        future_data.append([year, sector])

future_df = pd.DataFrame(future_data, columns=["Year", "Sector_Code"])
future_preds = model.predict(future_df.values)

future_df["Predicted_FDI"] = future_preds

# Save CSV
future_df.to_csv(os.path.join(output_dir, "ais_future_predictions.csv"), index=False)

# Graph
plt.figure()
plt.plot(future_preds)
plt.title("AIS Future Prediction")
plt.savefig(os.path.join(output_dir, "ais_future_graph.png"))
plt.close()

# ================================
# ✅ DONE
# ================================
print("✅ AIS Model completed successfully with all ais_ outputs!")
```
